It takes all that raw, messy "shit" recorded by the car’s iot sensors and the FIA's timing systems and is called telemetry here in fastf1

The Inputs: What the driver is doing (Throttle, Brake, Gear, DRS).

The Output: What the car is doing (Speed, RPM).

The Position: Where the car is doing it (Distance, RelativeDistance, Time).

The Map: Where the car is in 3D space (X, Y, Z coordinates).

```python
telemetry = nor_laps.get_telemetry()
print(telemetry.columns)
```

**Columns what and what it justifies:**

> DriverAhead: indicates traffic, Explains abnormal braking, throttle lifts, or speed loss, Helps separate driver limitation vs race context

> DistanceToDriverAhead: dirty air, can justify drivers pace drop without blaming the driver, Explains reduced corner speed or delayed throttle

> Time: Timestamp of telemetry, starting from 0 to when stint ended

> Distance: Absolute distance travelled since lap start (meters).

>RelativeDistance : Distance normalized from 0 → 1 across lap wrt other distances. Measures How connected or disconnected they are - despite their absolute distance from each other
>> why? Aligns laps of different lengths, for lap-to-lap comparison

>Speed: km/hr of the car

>RPM: Engine revolutions per minute. IT Shows power delivery, Gear selection logic

> nGear: Detects upshifts/downshifts, to know the type of feature in track atm, RPM–speed relationship

> Throttle: 0-100 throttle pedal position aka pressure.
Why:- Indicates intent (attack vs conserve), Shows corner exit behavior, Used for throttle zoning

> Brake: Boolean when braked or na indicator of braking points, Used to count braking events

>DRS (Drag Reduces areodynamically and driver have DRS over the front car...hence the speed benifit)

: its a boolean that explains sudden speed gain & throttle pressure,Validates straight-line behavior, prevents of not naming it as drivers skill.

>TyreLife : Number of laps on current tyre. Explains pace degradation

> Compound: Tyre compound (Soft/Medium/Hard). Contextualizes speed and braking (as: soft is initially position savy and hard takes time to warm up and last long hence pace.)

Allows fair comparison across stints as each compound has their own characteristics.

> Track Position X, Y, Z : Car coordinates in track reference frame.

```python
telemetry = []

for _, lap in data.iterlaps():
    lap_tel = lap.get_telemetry()
    lap_tel['LapNumber'] = lap['LapNumber']
    lap_tel['TyreLife'] = lap['TyreLife']
    lap_tel['Compound'] = lap['Compound']
    lap_tel['Stint'] = lap['Stint']
    telemetry.append(lap_tel)

telemetry = pd.concat(telemetry, ignore_index=True)

#took these from fastf1, additional stuff that needs to be added to telemetry []
```

**the _,**

the underscore _ is a throwaway variable name.

"A value is being returned here, but I don't plan on using it."

What happens if you don't use it?

If you wrote for lap in data.iterlaps(): your lap variable would actually become a tuple containing (index, data). When you then tried to call lap.get_telemetry(), the code would crash because a tuple doesn't have a get_telemetry method.

By using _, lap, you "unpack" the tuple immediately, putting the useless index in _ and the useful data in lap.

```python
#Convert booleans T/F to ints 1/0
telemetry["Brake"] = telemetry["Brake"].astype(int)
telemetry["DRS"] = telemetry["DRS"].astype(int)
```

```python
telemetry['Time'] = pd.to_timedelta(telemetry['Time'] ,errors="coerce").dt.total_seconds()

#as time is in string (0 days 00:58...) as its iot sensor data. Needs to make it into seconds by to_timedelta() into dt.total_seconds()
```

** New Colms and why they are needed:**

>

```python
telemetry['Speed_ms'] = (telemetry['Speed'] / 3.6).astype(float)
telemetry['dt'] = (telemetry['Time'].diff()).astype(float)
#whats diff() ? it takes the difference between the values for each row and, by default wrt the previous row

telemetry['Acceleration'] = telemetry['Speed_ms'].diff() / telemetry['dt']

# dt :- Time difference between samples. as Time is wrt time of the whole race, we need time_diff per lap type.
#Like for eg if we keep the diff b/w laps to 1: then time lap 0 = 0 then by lap 15 it would be 15 but time of lap 15 is actially 1
```

```python
print((telemetry['Time'] < 0).sum())
#checked no -ve time.
```

```python
telemetry['jerk'] = telemetry['Acceleration'].diff() / telemetry['dt']

# jerk :- if acce changes wrt time...it leads to jerking. This with / dt; the rate of change of acceleration over time.
```

```python
telemetry['speed_Delta'] = telemetry['Speed'].diff()

# speed_delta :- Raw speed change between samples, needed to Detects sudden speed loss/gain and be in sync with acceleration...

#btw: Proxy metric, not physical torque
```

```python
telemetry['Braking_event'] = telemetry["Brake"].diff().fillna(0) == 1

#means whenever braking changes True->False then do it 1->0
```

```python
conditions = [
    (telemetry["Throttle"] <= 20),
    (telemetry["Throttle"] > 20) & (telemetry["Throttle"] <= 70),
    (telemetry["Throttle"] > 70)
]

choices = ["Low", "Medium", "High"]

telemetry["Throttle_zone"] = np.select(conditions, choices, default="Unknown")

#throttle_zone :- categorizing throttle pressure outputs as Labels for easier intepreation and visualization of zones, Separates lift, modulation, full push
```

```python
condlist = [
    (telemetry['Speed'] <= 50),
    (telemetry['Speed'] > 50) & (telemetry['Speed'] <= 130),
    (telemetry['Speed'] > 130) & (telemetry['Speed'] <= 210),
    (telemetry['Speed'] > 210) & (telemetry['Speed'] <= 290),
    (telemetry['Speed'] > 290)
]

choicelist = [
    "Slow",
    "Slow-medium",
    "High-medium",
    "High",
    "Extreme-high"
]

telemetry['Speed_zone_all'] = np.select(condlist, choicelist, default="Unknown")

# Speed zones defined:
# <=50: hairpins, pit
# 50–130: tight corners
# 130–210: fast bends
# 210–290: high speed-slight traffic
# >290: straight-line / DRS
```

```python
import numpy as np

#making braking intensity to DETECt how fast braked to categorize a reason
braking_data = (
    (telemetry['Acceleration'] < 0) &
    (telemetry['Speed'] > 30) &
    (telemetry['Throttle'] < 5)
)

#coasting, lift-and-roll, low-speed
telemetry['Brake_Intensity'] = (-telemetry['Acceleration']).where(braking_data, 0)


#Why - ? we take -ve coz we are calc braking. While braking the acc is negative as speed drops. hence we multiple by -ve...so if its +ve it becomes -ve and hence cannot be 0-100% hence condition is false. Else if -3.4 acce, means braking and when we x by -1 then it ebcomes +ve and can be makde into a %that we need
# 0  is fall back value, when the % isnt positive (after x by -1) then value of % is to be 0


print(f"Max: {telemetry['Brake_Intensity'].max()}")
print(f"Min (above 0): {telemetry['Brake_Intensity'][telemetry['Brake_Intensity'] > 0].min()}")

# Define categories
condlist = [
    (telemetry['Brake_Intensity'] > 0) & (telemetry['Brake_Intensity'] <= 10),
    (telemetry['Brake_Intensity'] > 10) & (telemetry['Brake_Intensity'] <= 38),
    (telemetry['Brake_Intensity'] > 38)
]

choicelist = ["Low", "Mid", "High"]

telemetry['Braking_intensity_zone'] = np.select(condlist, choicelist, default="None")

telemetry['Brake_Intensity'] = telemetry['Brake_Intensity'].where(
    telemetry['Braking_event'], 0
)
```

**The numpy.select() function-**

is used to construct a new array by choosing elements from a choicelist based on a corresponding list of boolean conditions (condlist)

numpy.select(condlist, choicelist, default=0)

```python
telemetry["Gear_Change"] = telemetry["nGear"].diff().fillna(0)

# whenever gear is changed aka different. Counts gear shifts
```

**Why use -1 instead of 0?**

first bin starts at -1. This is a common trick to ensure that a throttle reading of exactly 0 gets included in the "Low" category. If you started at 0, the code might exclude the number 0 depending on the default settings.

```python
telemetry['Engine_load'] = telemetry['RPM'] * telemetry['Throttle']

# load is the product of throttle atm * engine revolution per min.

# engine_load : Proxy for engine stress and Combines intent + engine state as to see how much 'pressure' is on the engine due to everything load_rendering.
```

```python
telemetry['TB_overlap'] = (telemetry['Throttle'] > 0) & (telemetry['Brake'] > 0)

#Throttle-Brake Overlap :- whenever they both overlap.
```

Why is TB_Overlap Good Sign for drivers COmpetence?

>> When a driver jumps directly from throttle to brake, the car experiences a very abrupt weight transfer:

Rear loses load too fast

Front gains load too fast

The balance snaps instead of transitions

That snap = instability.

So good drivers overlap throttle and brake briefly to control how fast that weight transfer happens.

### Overlap isn’t wasted input; it’s how skilled drivers control weight transfer instead of jerking the car.

A smooth transition from throttle to brake naturally creates an overlap.
A strict switch between the two causes jerk and instability, whereas a smooth overlap allows a more neutral load transfer and a stable corner entry.
